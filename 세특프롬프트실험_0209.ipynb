```python
def topic_generator(keyword, criteria):
  prompt = ChatPromptTemplate(
    messages=[
        SystemMessagePromptTemplate.from_template("""
    <goal>
    * As an AI specialized in assisting high school students with curriculum exploration activities, you are helping a high school student who has a keen interest in 'major keyword' and is currently studying subject
    * meeping the 'subject achievement criteria', recommend four research topics related to 'major keyword' that this high school student can explore
    * Please provide brief explanations for the recommendations from the perspective of career development.

    <output format>
    * generated research topic
    * '::' should be added at the end criteria. after that, explaination has to be come up.
    * each result should be paired with generated topic::explain in array.
    "생태계와 바이오에너지의 연관성 이해하기::생태계의 에너지 흐름을 이해하고, 에너지 흐름을 물질 순환과 비교하여 차이를 설명할 수 있습니다. 또한, 생물다양성의 중요성을 이해하고 생물다양성 보전 방안을 토의하는 부분에서 바이오에너지와 생태계의 상호작용을 고려할 수 있습니다."]

     <example input>
     keyword: 반도체 공급망 문제
     criteria: 물질의 속성과 결합의 이해


     <example output>
     ["반도체 소자의 물리적 성질과 결합의 영향:: 이 주제를 통해 반도체 소자에 사용되는 물질들의 물리적 성질과 결합의 종류를 탐구하고, 이러한 특성이 반도체의 동작 및 성능에 어떻게 영향을 미치는지 이해할 수 있습니다.",
      "지속 가능한 반도체 소자 제조를 위한 물질 선택:: 반도체 소자의 제조 과정에서 사용되는 물질에 대한 환경적 영향을 고려하면서, 더 지속 가능하고 친환경적인 소자 제조를 위한 물질 선택에 관한 연구를 수행할 수 있습니다.",
      "화석 연료 의존도와 반도체 산업 간의 관계 분석:: 반도체 공급망 문제와 관련하여 화석 연료 의존도가 반도체 산업에 미치는 영향을 분석하고, 신재생 에너지 전환의 필요성에 대한 논의를 제시할 수 있습니다.",
      "에너지 효율적인 반도체 소자의 설계와 제조:: 이 주제를 통해 반도체 소자의 에너지 효율적인 설계 및 제조에 대한 기술적 측면을 탐구하고, 화석 연료 소모를 줄이며 에너지 효율을 향상시키는 방안을 연구할 수 있습니다."]
       """),

        HumanMessagePromptTemplate.from_template("""please
                                                    <input information>
                                                    major keyword: {keyword}
                                                    criteria: {criteria}

                                                    <output>
                                                    """)
        ],
    input_variables = ['keyword', 'criteria'],
    # partial_variables = {'format_instructions':format_instructions}
    # partial_variables={"format_instructions": format_instructions}
  )
  _input = prompt.format_prompt(keyword = keyword, criteria=criteria)

  with get_openai_callback() as cb:
    print('작업 들어갑니데이')
    # print('major', major)
    print('criteria', criteria)
    print('keyword', keyword)

    start = time.time()
    output = llm.invoke(_input.to_messages())
    end = time.time()
    print(f'answer: {output.content}')
    print(cb)
    print(f'{end - start} 경과')
  return output.content
```

```python
response_schemas= [
    ResponseSchema(name="ig", description="content of introduction part"),
    ResponseSchema(name="bg", description="content of body part"),
    ResponseSchema(name="cg", description="content of conclusion part"),
    ]
output_parser = StructuredOutputParser.from_response_schemas(response_schemas)
format_instructions = output_parser.get_format_instructions()


def main_guideline_generator(topic, major, explaination):
  prompt = ChatPromptTemplate(
    messages=[
        SystemMessagePromptTemplate.from_template("""
    <goal>
    * As an AI specialized in assisting high school students with curriculum exploration activities, you are helping a high school student who has a keen interest in 'major keyword' and is currently studying subject
    * A high school student interested in 'major' is preparing to write a research report.
    * provide guidelines for writing a research report with topic, major and keyword

    <output format>
    * Should be devided into 3 parts as 'introduction', 'body', 'conclusion'
    * put the subtopic in '<>' and put the context after '\\n* ' and put the next subtopic in '<>' after '\\n\\n'
    * example: <subtopic1>\\n* context1\\n* context2\\n\\n<subtopic2>\\n* context3\\n* context4

    <essay introduction guide>
    * Provide assistance to explain the topic
    * Provide guideline to write the goal of research

    <essay body guide>
    * Provide detailed contents of introduction's guidance
    * Provide guidance for conducting research or analysis activities related to fundamental theoretical knowledge concerning the topic.
    * Create guidelines for investigating the market, industry practices, or current industry status related to the topic's impact assessment.
    * Develop guidelines for researching based on cases related to the topic.

    <essay conclusion guide>
    * Offer guidance on contemplating the outlook of the industry or specific career paths to help users consider future prospects.
    * Provide guidelines on how to effectively structure and present overall research findings.
    * Supply guidance on suggesting future research directions to users.
    {format_instructions}

    <example input>
    major: 반도체공학
    keyword: 반도체 공급망 문제
    topic: 지속 가능한 반도체 소자 제조를 위한 물질 선택

    <example output>
    "{{"ig": "<주제 소개>\\n반도체 소자의 생산과정에서 사용되는 물질의 선택이 환경에 미치는 영향을 다룹니다.\\n\\n<연구 목적>\\n더 지속 가능하고 환경 친화적인 반도체 소자 제조를 위한 물질 선택에 관한 연구의 목적을 설명합니다.",
    "bg":"<환경 영향 분석>\\n* 과거의 반도체 소자 생산에서 사용된 물질들이 환경에 미치는 부정적인 영향을 조사합니다.\\n* 최근의 환경 지속 가능성에 대한 요구사항을 바탕으로, 새로운 물질의 선택이 어떻게 기존 방법을 개선할 수 있는지 분석합니다.\\n\\n<글로벌 반도체 시장 현황>\\n* 글로벌 반도체 시장에서 사용되는 주요 소재 및 해당 물질의 생산과 사용에 따른 환경 영향을 조사합니다.\\n* 각 국가의 환경 규제 및 정책이 반도체 소자 제조에 미치는 영향을 분석합니다.\\n\\n<미국의 팹4 제안과 의미>\\n*미국이 제안한 팹4에 참여함으로써 어떤 물질이 사용되며, 이로 인해 어떤 환경적 효과가 발생할지 조사합니다.\\n*한국의 반도체 산업에 미치는 영향 및 한국의 대응 전략을 분석합니다.",
    "cg":"<중장기 전망과 대응 전략>\\n* 반도체 소자 제조에서의 물질 선택이 지속 가능성과 환경에 미치는 영향에 대한 중장기적인 전망을 제시합니다.\\n* 한국이 지속 가능한 반도체 소자 생산을 위해 어떤 전략을 채택해야 하는지 제언합니다.\\n\\n<전체적인 연구 결과 강조>\\n* 연구에서 얻은 주요 결과와 환경에 대한 고려사항을 간략하게 요약합니다.\\n* 반도체 소자 제조에 관한 물질 선택이 환경 지속 가능성에 어떻게 기여할 수 있는지 강조합니다.\\n\\n<향후 연구의 방향 제시>\\n* 환경 친화적인 물질의 개발 및 적용에 관한 추가적인 연구가 필요한 이유를 제시합니다.\\n* 더 나아가, 반도체 소자 생산에서의 물질 선택이 다양한 기술적, 경제적, 환경적 측면에서 어떻게 협업하여 진행되어야 하는지 논의합니다."}}"
       """),

        HumanMessagePromptTemplate.from_template("""
                                                    <input information>
                                                    major: {major}
                                                    major keyword: {keyword}
                                                    topic: {topic}

                                                    <output>
                                                    """)
        ],
    input_variables = ['keyword', 'topic', 'major'],
    # partial_variables = {'format_instructions':format_instructions}
    partial_variables={"format_instructions": format_instructions}
  )
  _input = prompt.format_prompt(major = major, keyword = keyword, topic=topic)

  with get_openai_callback() as cb:
    print('작업 들어갑니데이')
    print('major', major)
    print('topic', topic)
    print('keyword', keyword)

    start = time.time()
    output = llm.invoke(_input.to_messages())
    end = time.time()
    print(f'answer: {output.content}')
    print(cb)
    print(f'{end - start} 경과')
  return output.content
```
